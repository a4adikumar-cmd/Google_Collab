INSTANCE METHOD

```python
#1
class car:
    def display_brand_name(self):     #instance method
        print("brand is Toyota")

s1 = car()                            #object created
s1.display_brand_name()               #instance method calling
```

```python
#2
class clinic:
    def patient_details(self, name, age):       #instance method 1
        self.name = name
        self.age = age

    def display_details(self):                  #instance method 2
        print("Name:", self.name)
        print("Age:", self.age)


s1 = clinic()                                   #object created

s1.patient_details("Aditya", 22)                #passing argumemts to instance method 1
s1.display_details()                            #instance method 2 calling
```

```python
#3
class employee:
    def details(self, name, salary):          #instance method 1
        self.name = name
        self.salary = salary

    def display_details(self):                    #instance method 2
        print("Employee:", self.name)
        print("Salary:", self.salary)


e1 = employee()                                   #object 1 created
e2 = employee()                                   #object 2 created

e1.details("Rahul", 50000)                   #instance method 1 calling using object 1
e2.details("Aman", 60000)                    #instance method 1 calling using object 2

e1.display_details()
e2.display_details()
```

```python
#4
class bank_account:
    def set_balance(self, amount):                      #instance method 1
        self.balance = amount

    def deposit(self, amount):                          #instance method 2
        self.balance += amount
        self.show_balance()

    def show_balance(self):                             #instance method 3
        print("Current Balance:", self.balance)


b1 = bank_account()                                    #object created

b1.set_balance(1000)
b1.deposit(500)
```

```python
#5
class rectangle:
    def dimensions(self, length, width):                   #instance method 1
        self.length = length
        self.width = width

    def calculate_area(self):                              #instance method 2
        area = self.length * self.width
        print("Area of rectangle is", area)


r1 = rectangle()                                               #object created

r1.dimensions(10, 5)
r1.calculate_area()
```

```python
#6
class shopping_cart:
    def create_cart(self):                    #instance method 1
        self.cart = []

    def add_item(self, product):              #instance method 2
        self.cart.append(product)
        print(product, "added")

    def remove_item(self, product):           #instance method 3
        if product in self.cart:
           self.cart.remove(product)
        print(product, "removed")

    def show_cart(self):
        print("Products in cart:", self.cart)


cart = shopping_cart()

cart.create_cart()

cart.add_item("Shirt")
cart.add_item("Soy protein")
cart.remove_item("spects")

cart.show_cart()
```

CLASS METHOD

```python
# Ex - 1
class Vehicle:
    color = 'white'                 #class variable

    @classmethod
    def changeColor(cls, color):    #class method
        cls.color = color           #accessing class variable
        print(f'Color of vehicle is changed to "{color}"')

Vehicle.changeColor("Red")          #calling class method
```

```python
# Ex - 2
class College:
  Dept = "MCA"

  @classmethod
  def displayDept(y):                 #class method
    print(f"Department is {y.Dept}")  #will display class variable

College.displayDept()                 #calling class method
```

```python
# Ex - 3
class Currency:
  Dollar_to_Inr = 86;

  @classmethod
  def Convert_Dollar_to_Inr(C, dollar):    # C is to take class name as an argument
    return dollar * C.Dollar_to_Inr        # Operates using class variable

dollar = int(input("Enter the amount in Dollars: "))
print(dollar,"dollars cost", Currency.Convert_Dollar_to_Inr(dollar)," INR.")
```

```python
# Ex - 4
class Account:
    Acc_Bal = 0

    @classmethod
    def withdraw(cls, amount_to_withdraw):                     #1st class method
      if amount_to_withdraw > cls.Acc_Bal:
        print("Insufficient Balance")
      else:
        cls.Acc_Bal = cls.Acc_Bal - amount_to_withdraw
        print("Withdrawal successful")
        print("Account Balance:",cls.Acc_Bal)


    @classmethod                                              #2nd class method
    def Deposit(cls, amount_to_deposit):
        if amount_to_deposit < 0:
            print("Invalid amount")
        else:
            cls.Acc_Bal = cls.Acc_Bal + amount_to_deposit
            print("Deposit successful")
            print("Account Balance:",cls.Acc_Bal)

Account.withdraw(500)
Account.Deposit(1000)
```

```python
# Ex - 5
class Bank:
    interest_Rate = 6.5

    @classmethod                                                   #1st class method
    def show_Rate(cls):
        print("Interest Rate:", cls.interest_Rate, "%")

    @classmethod                                                   #2nd class method
    def increase_Rate(cls, value):
        cls.interest_Rate += value

    @classmethod                                                   #3rd class method
    def set_Rate(cls, rate):
        cls.interest_Rate = rate


Bank.show_Rate()

Bank.increase_Rate(0.5)
Bank.show_Rate()

Bank.set_Rate(7.2)
Bank.show_Rate()
```

```python
# Ex - 6
class Car:
    color = "White"
    brand = "Toyota"

    @classmethod
    def about_car(cls):                        #1st class method
        print("Brand:", cls.brand)
        print("Color:", cls.color)

    @classmethod                               #2nd class method
    def change_Color(cls, new_Color):
        cls.color = new_Color
        print("Color of car is changed to", cls.color)

    @classmethod                               #3rd class method
    def change_Brand(cls, new_Brand):
        cls.brand = new_Brand
        print("Brand of car is changed to", cls.brand)

Car.about_car()
print()
Car.change_Color("Red")
Car.change_Brand("Honda")
print()
Car.about_car()
```

STATIC METHOD

```python
#1
class Calculator:

    @staticmethod
    def add(a, b):                                 #static method
        return a + b

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print(f"{a} + {b} = {Calculator.add(a,b)}")        #static method calling inside print()
```

```python
#2
class Number:

    @staticmethod
    def is_Even(number):                   #static method
        return number % 2 == 0

a = int(input("Enter a number: "))
if Number.is_Even(a):                     #static method calling in condition
    print(f"{a} is even")
else:
    print(f"{a} is odd")
```

```python
#3
class String:

    @staticmethod
    def count_Vowels(text):              #static method
        count = 0
        for ch in text.lower():
            if ch in "aeiou":
                count += 1
        return count

s = input("Enter a string: ")
print(f"Number of vowels in {s} is {String.countVowels(s)}")  #static method calling inside print()
```

```python
#4
class Email:

    @staticmethod
    def is_Valid(email):                          #static method
        if email.endswith("@gmail.com"):
            print(email, "is a valid mail.")
        else:
            print(email, "is an invalid mail.")

e = input("Enter an Email: ")
Email.is_Valid(e)                                 #static method calling
```

```python
#5
class CheckString:

    @staticmethod
    def count_characters(text):                   #static method
        print("Total Characters:", len(text))

    @staticmethod
    def convert_to_upper(text):
        print("Uppercase:", text.upper())

text = input("Enter a string: ")
CheckString.count_characters("python")           #static method calling
CheckString.convert_to_upper("hello world")      #static method calling
```

```python
#6
class calculator:

    @staticmethod
    def add(num1, num2):                                 #static method
        print(f"{num1} + {num2} = {num1 + num2}")

    @staticmethod
    def multiply(num1, num2):
        print(f"{num1} * {num2} = {num1 * num2}")

calculator.add(15, 25)                                   #static method calling
calculator.multiply(6, 8)
```
